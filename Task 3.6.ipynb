```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
```

# Генерация данных

```python
# # Устанавливаем начальное значение для генератора случайных чисел для обеспечения воспроизводимости
np.random.seed(26) 

# Генерируем массив из 1000 случайных чисел для группы A, которые следуют нормальному распределению
# с средним (loc) 50 и стандартным отклонением (scale) 10
group_a = np.random.normal(loc=25, scale=8, size=1000)

# Генерируем массив из 1000 случайных чисел для группы B
group_b = np.random.normal(loc=24, scale=12, size=1000)
```

# Визуализация данных

```python
# Устанавливаем размер фигуры
plt.figure(figsize=(8, 6))

# Строим гистограмму для группы A
plt.hist(group_a, bins=30, alpha=0.5, label='Группа A')

# Строим гистограмму для группы B
plt.hist(group_b, bins=30, alpha=0.5, label='Группа B')
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.title('Гистограмма двух контрольных групп')
plt.legend()
plt.show()
```

# t-тестирование

```python
alpha = 0.05

t_stat, p_value = ttest_ind(group_a, group_b)
print(f'Значение p-value: {p_value: .4f}')

if p_value > alpha:
    print('Статистичиески значимых различий между группами не обнаружено.' )
else:
    print('Обнаружены статистические значимые различия между группами.' )
```
