## Условный оператор if...else

**2. Напишите программу, которая на вход будет получать количество отработанных сотрудником часов и показатель производительности (воспользуйтесь методом `input`, изучите его самостоятельно, он предназначен для того, чтобы значения в него закладывались после выполнения ячейки с кодом). Количество отработанных часов должно быть не менее 40, а показатель производительности положительным. В случае удовлетворения этих условий необходимо посчитать эффективность работы сотрудника по формуле `время` * `производительность`, иначе же необходимо вывести замечание сотруднику**

Пример результата:

```
Введите количество отработанных часов: 42
Введите показатель производительности: 5
Эффективность работы сотрудника составляет 210.0 единиц в час
```
```
Введите количество отработанных часов: 38
Введите показатель производительности: 5
Количество часов должно быть больше 40
```

````python
try:
    print('Введите количество отработанных часов')
    hour=input()
    print('Введите показатель производительности')
    proisv=input()
    res=''
    if(int(hour)>=40 and int(proisv)>=0):
        res+=f'Эффективность работы сотрудника составляет: {int(hour)*int(proisv)} единиц в час'
    if int(proisv)<0:
        res+=f'Производительность должна быть положительна '
    if int(hour)<40 :
        res+='Количество отработанных часов должно быть больше 40 '
    print(res)
except:
    print('Ошибка ввода')
````

##  Цикл For

**3. Напишите программу, которая после выполнения выдает список значений, удовлетворяющих теореме Пифагора `(Пифагоровы тройки)`.
Переберите все возможные комбинации, где сумма всех чисел будет меньше 50**

````python
pt=[]
for i in range(1,50):
    for j in range(i+1,50):
        k=(i**2+j**2)**0.5
        if int(k)==k and i+j+k<50:
            pt+=[[i,j,int(k)]]
pt
````

##  Функции

**4. Напишите функцию для поиска наименьшего общего кратного (НОК) двух чисел. Функция должна принимать на вход два параметра и возвращать наименьшее общее кратное этих чисел**

````python
ab=input('Введите 2 целых числа')
a, b=[int(c) for c in ab.split()]
NOK=a if a>=b else b
while NOK %a+NOK%b>0:
    NOK+=1
NOK
````
